### Import libraries

## Creating the game logic

### Game Class
-The `Game` class encapsulates the logic of the Rock Paper Scissors game. 
-It provides methods for getting the computer's choice, getting the user's choice, determining the winner, and playing the game.

#### `get_computer_choice()`
- **Static Method**: This method is marked with `@staticmethod` because it doesn't rely on any instance-specific data. 
- It simply returns a random choice of "Rock", "Paper", or "Scissors" for the computer.

#### `get_user_choice()`
- **Static Method**: Similar to `get_computer_choice()`, this method is marked with `@staticmethod` because it doesn't require any instance-specific data.
- It prompts the user for input and returns their choice after validating it against the list of valid choices.

#### `get_winner(user_choice, computer_choice)`
- **Static Method**: Again, marked with `@staticmethod` because it operates on the input choices without needing access to instance-specific data.
- It compares the user's and computer's choices to determine the winner based on the rules of Rock Paper Scissors.

#### `play()`
- **Class Method**: This method is marked with `@classmethod` because it operates on the class itself rather than an instance of the class. 
- It simulates playing the game, alternating between getting the user's choice and generating the computer's choice until one side wins three rounds or five rounds have been played.

### Explanation of Logic

- Each button representing a game choice ("Rock", "Paper", "Scissors") has a command attached to it, which calls the `play_round` function with the corresponding choice.
- Inside `play_round`, the computer's choice is determined using the `get_computer_choice` method from the `Game` class, and the winner is determined using the `get_winner` method.

```python
"""Rock Paper Scissors - Death Match """


# Import necessary libraries
import random 
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

"""Game Logic coding prior to incorporating it into GUI"""
# Define the Game class to encapsulate game logic
class Game:
    # Method to get computer's choice
    @staticmethod
    def get_computer_choice():
        choices = ["Rock", "Paper", "Scissors"]
        return random.choice(choices)

    # Method to get user's choice
    @staticmethod
    def get_user_choice():
        choices = ["Rock", "Paper", "Scissors"]
        while True:
            user_choice = input("Your move: ").capitalize()
            if user_choice in choices:
                return user_choice
            else:
                print("Invalid input!")

    # Method to determine the winner
    @staticmethod
    def get_winner(user_choice, computer_choice):
        if user_choice == computer_choice:
            return "It's a draw!"
        elif (user_choice == "Rock" and computer_choice == "Scissors") or \
             (user_choice == "Paper" and computer_choice == "Rock") or \
             (user_choice == "Scissors" and computer_choice == "Paper"):
            return "User wins!"
        else:
            return "Computer wins!" 

    # Method to play the game
    @classmethod
    def play(cls):
        user_wins = 0
        computer_wins = 0
        rounds_played = 0

        # Play until one side wins 3 rounds or 5 rounds have been played
        while user_wins < 3 and computer_wins < 3 and rounds_played < 5:
            computer_choice = cls.get_computer_choice()
            user_choice = cls.get_user_choice()
            print("Computer's choice:", computer_choice)
            print("User's choice:", user_choice)

            winner = cls.get_winner(user_choice, computer_choice)
            print(winner)

            if winner == 'User wins!':
                user_wins += 1
            elif winner == 'Computer wins!':
                computer_wins += 1
            
            rounds_played += 1
            print(f"Score: User {user_wins} : Computer {computer_wins}")

        # Print the overall winner
        if user_wins > computer_wins:
            print("Congratulations! You won the game!")
        else:
            print("Computer wins the game!")


"""GUI: Now incorporating the game logic into the Tkinter GUI to create an interactive game"""
# Function to create and run the Tkinter GUI
def play_game():
    root = tk.Tk()
    root.title("Rock Paper Scissors")

    #Code to incorporate a background image into the GUI
    background_image = Image.open("rps_background.jpg")
    background_photo = ImageTk.PhotoImage(background_image)
    background_label = tk.Label(root, image=background_photo)
    background_label.place(x=0, y=0, relwidth=1, relheight=1)

    # Function to play a round when a button is clicked
    def play_round(user_choice):

        nonlocal user_wins, computer_wins, rounds_played
        computer_choice = Game.get_computer_choice()
        computer_choice_label.config(text=f"Computer plays: {computer_choice}")
        winner = Game.get_winner(user_choice, computer_choice)
        result_label.config(text=winner)

        if winner == 'User wins!':
            user_wins += 1
        elif winner == 'Computer wins!':
            computer_wins += 1
        
        rounds_played += 1
        rounds_played_label.config(text= f"Round: {rounds_played}")
        score_label.config(text= f"Score:  User {user_wins}  -  {computer_wins} Computer")

        # If one side wins 3 rounds, show game over message and reset
        if user_wins == 3 or computer_wins == 3:
            if user_wins == 3:
                messagebox.showinfo("Game Over", "You won the game!\nClick OK to play again")
            else:
                messagebox.showinfo("Game Over", "Computer won the game!\nClick OK to play again")
            reset_game()
    
    # Function to reset the game
    def reset_game():
        nonlocal user_wins, computer_wins, rounds_played
        user_wins = 0
        computer_wins = 0
        rounds_played = 0
        computer_choice_label.config(text="")
        rounds_played_label.config(text="")
        result_label.config(text="")
        score_label.config(text="")
    
    # Initialize variables
    user_wins = 0
    computer_wins = 0
    rounds_played = 0

    # Create buttons for choices
    choices = ["Rock", "Paper", "Scissors"]
    for i, choice in enumerate(choices): 
        button = tk.Button(root, text=choice, command=lambda ch=choice: play_round(ch))
        button.place(relx=0.5, rely=0.3 + 0.1*i, anchor='center')
    
    
    # Labels to display computer choice, result, and score, and their placement in the GUI
    
    computer_choice_label = tk.Label(root,text= "", bg="white")
    computer_choice_label.place(x=300, y=100)
   
    rounds_played_label = tk.Label(root, text= "", bg="white")
    rounds_played_label.place(x=300, y=125)
   
    result_label = tk.Label(root, text= "", bg="white")
    result_label.place(x=300, y=150)
   
    score_label = tk.Label(root, text = "", bg="white")
    score_label.place(x=300, y=175)

    root.mainloop()

# Call the game class to begin the game
play_game()
```
